```python
import matplotlib.pyplot as plt
import numpy as np

# Old values for F1 scores
old_entity_f1 = 84.63
old_entity_f1_std = 0.25
old_relation_f1 = 67.41
old_relation_f1_std = 0.92

# New values for F1 scores
new_entity_f1 = 85.21
new_entity_f1_std = 0.0
new_relation_f1 = 69.18
new_relation_f1_std = 0.0

# Data for plotting
categories = ['Entity F1', 'Relation F1']
old_values = [old_entity_f1, old_relation_f1]
new_values = [new_entity_f1, new_relation_f1]
old_errors = [old_entity_f1_std, old_relation_f1_std]
new_errors = [new_entity_f1_std, new_relation_f1_std]

# Set the bar width and positions
bar_width = 0.3
index = np.arange(len(categories))

# Plotting the bars
fig, ax = plt.subplots(figsize=(8, 6))

# Choosing light pastel colors from a palette
pastel_pink = 'red'
pastel_lavender = 'blue'  # Light lavender

# Plotting bars with chosen pastel colors
bars1 = ax.bar(index - bar_width/2, old_values, bar_width, yerr=old_errors, label='Old Values', capsize=5, color=pastel_pink, edgecolor='black', linewidth=1.5)
bars2 = ax.bar(index + bar_width/2, new_values, bar_width, yerr=new_errors, label='New Values', capsize=5, color=pastel_lavender, edgecolor='black', linewidth=1.5)

# Adding data labels to the bars
def add_data_labels(bars):
    for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom',
                    fontsize=12, color='black')

add_data_labels(bars1)
add_data_labels(bars2)

# Enhancing the graph appearance
ax.set_xlabel('F1 Score Category', fontsize=12, fontweight='bold')
ax.set_ylabel('F1 Score (%)', fontsize=12, fontweight='bold')
ax.set_title('BERT MODEL: Comparison of Entity and Relation F1 Scores (Old vs New)', fontsize=14, fontweight='bold')
ax.set_xticks(index)
ax.set_xticklabels(categories, fontsize=12)
ax.set_yticks(np.arange(65, 90, 5))
ax.tick_params(axis='y', labelsize=12)
ax.legend(fontsize=12, loc='upper right')

# Adding grid lines for better readability
ax.yaxis.grid(True, linestyle='--', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import torch
from transformers import BertTokenizer
from models.spert.spert.models import SpERT

# Load the tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Define model arguments
cls_token = 101  # Add your value for the cls_token
num_relation_types = 18
num_entity_types = 14
size_embedding = 25
prop_drop = 0.3

# Load the SpERT model with BERT
model = SpERT.from_pretrained(
    'bert-base-uncased',
    cls_token=cls_token,
    num_relation_types=num_relation_types,
    num_entity_types=num_entity_types,
    size_embedding=size_embedding,
    prop_drop=prop_drop
)

# Load the model weights (set strict=False to ignore missing/extra keys)
model_path = 'log/2025-03-06_23:24:23.514424_spert/iter_0/best_model/pytorch_model.bin'
model_load = torch.load(model_path)

# Resize model's word embeddings to the checkpoint size (if necessary)
model.bert.embeddings.word_embeddings = torch.nn.Embedding(28996, 768)  # Set the size to the one from the checkpoint

# Load the state dict, while ignoring mismatches in size (strict=False)
missing_keys, unexpected_keys = model.load_state_dict(model_load, strict=False)

# Print missing and unexpected keys
print(f"Missing keys: {missing_keys}")
print(f"Unexpected keys: {unexpected_keys}")

model.eval()  # Set to evaluation mode

# Function to create padded entity data (for demonstration purposes)
def create_padded_entity_data(inputs, num_entities=6):
    batch_size = inputs['input_ids'].size(0)  # Batch size
    sequence_length = inputs['input_ids'].size(1)  # Length of the input sequence
    
    # Padding entity-related data to match the sequence length
    entity_sizes = torch.ones(batch_size, num_entities).long()  # Assume 6 entities per example (dummy)
    entity_spans = torch.zeros(batch_size, num_entities, 2).long()  # Start and end for 6 entities
    entity_sample_masks = torch.ones(batch_size, num_entities).long()  # All entities are valid
    entity_masks = torch.ones(batch_size, sequence_length).long()  # Adjusted to the sequence length
    
    # Create context mask (for attention purposes)
    context_masks = inputs['attention_mask']
    
    # Pad entity sizes, spans, and masks to match the sequence length if needed
    if num_entities < sequence_length:
        padding_size = sequence_length - num_entities
        entity_sizes = torch.cat([entity_sizes, torch.zeros(batch_size, padding_size).long()], dim=1)
        entity_spans = torch.cat([entity_spans, torch.zeros(batch_size, padding_size, 2).long()], dim=1)
        entity_sample_masks = torch.cat([entity_sample_masks, torch.zeros(batch_size, padding_size).long()], dim=1)
    
    return entity_sizes, entity_spans, entity_sample_masks, entity_masks, context_masks

# Function for inference
def infer(text):
    # Tokenize the input text
    inputs = tokenizer(text, return_tensors='pt', padding=True, truncation=True, max_length=512)
    
    # Move model and inputs to the same device (GPU if available)
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)
    inputs = {key: value.to(device) for key, value in inputs.items()}
    
    # Get the padded entity-related data
    entity_sizes, entity_spans, entity_sample_masks, entity_masks, context_masks = create_padded_entity_data(inputs)
    
    # Check the shapes of the tensors before passing them to the model
    print(f"entity_sizes shape: {entity_sizes.shape}")
    print(f"entity_spans shape: {entity_spans.shape}")
    print(f"entity_sample_masks shape: {entity_sample_masks.shape}")
    print(f"entity_masks shape: {entity_masks.shape}")
    print(f"context_masks shape: {context_masks.shape}")
    
    # Pass inputs along with the additional entity-related data
    with torch.no_grad():  # Disable gradient computation for inference
        outputs = model(
            encodings=inputs['input_ids'], 
            context_masks=context_masks, 
            entity_masks=entity_masks,
            entity_sizes=entity_sizes, 
            entity_spans=entity_spans, 
            entity_sample_masks=entity_sample_masks, 
            inference=True
        )
    
    # Check what the model returns
    print(f"Model outputs: {outputs}")
    
    # The model output might be a tuple, so we extract the logits from it.
    if isinstance(outputs, tuple):
        logits = outputs[0]  # Assuming the first element of the tuple is logits
    else:
        logits = outputs.logits  # In case it's a model output object
    
    # Assuming this is a classification task, you can use argmax for the prediction
    prediction = torch.argmax(logits, dim=-1)  # Get the index of the maximum value in logits
    
    # If logits have shape [batch_size, num_classes], we use .item() for scalar prediction
    if prediction.dim() == 0:  # Only one prediction for the batch
        return prediction.item()
    else:
        # If there are multiple predictions, we might want to return the index of the max value per sample
        return prediction.squeeze().tolist()  # Convert to list if there are multiple predictions

# Example usage
text_input = "Shares of tesla dropped 15% over the last quarter."
prediction = infer(text_input)
print(f"Prediction: {prediction}")
```

```python
import json
import random
import torch
from transformers import BertTokenizer, BertForTokenClassification

# Load the tokenizer and model from the checkpoint
tokenizer = BertTokenizer.from_pretrained('bert-base-cased')  # Using 'bert-base-cased' as per your config
model = BertForTokenClassification.from_pretrained(
    'log/2025-03-07_02:22:50.879393_spert/iter_0/best_model',
    state_dict=torch.load('log/2025-03-07_02:22:50.879393_spert/iter_0/best_model/pytorch_model.bin')
)

model.eval()  # Set model to evaluation mode

# Load labels (entities and relations) from the correct JSON file
def load_labels(json_file_path):
    with open(json_file_path, 'r') as f:
        data = json.load(f)
    
    # Extract entities and relations
    entity_labels = {key: value['short'] for key, value in data['entities'].items()}
    relation_labels = {key: value['short'] for key, value in data['relations'].items()}
    
    return entity_labels, relation_labels

# Load the fire_rules.json for relation extraction rules
def load_rules(rules_file_path):
    with open(rules_file_path, 'r') as f:
        data = json.load(f)
    return data['rules']

# Load entity labels, relation labels, and rules dynamically
entity_labels, relation_labels = load_labels('fire/data/fire_types.json')  # Ensure this is the correct file path
rules = load_rules('fire/supplemental_resources/fire_rules.json')  # This file is for relation rules

# To generate random test example
def generate_random_test_example():
    test_examples = json.load(open('./fire/data/fire_test.json'))
    random_example = [random.choice(test_examples)]
    test_path = './test.json'
    with open(test_path, 'w') as f:
        json.dump(random_example, f)

    return random_example

# Function to perform NER (Named Entity Recognition) inference
def get_named_entities(sentence):
    inputs = tokenizer(sentence, return_tensors="pt", truncation=True, padding=True)
    outputs = model(**inputs).logits
    predictions = torch.argmax(outputs, dim=-1)

    # Convert token predictions to named entities (assuming labels 0 for "O" and other numbers for entity classes)
    tokens = tokenizer.convert_ids_to_tokens(inputs["input_ids"][0])
    entities = []

    for token, prediction in zip(tokens, predictions[0]):
        if prediction != 0:  # 0 is usually for non-entity, adjust if your model uses different labels
            entities.append((token, entity_labels.get(str(prediction.item()), 'Unknown')))
    
    return entities

# Function to perform RE (Relation Extraction) inference based on rules
def get_relations_from_rules(entities):
    relations = []

    # Loop through the rules to find valid relations
    for rule in rules:
        head_entity = rule['head']
        tail_entity = rule['tail']
        valid_relations = rule['relations']

        # Find entities of the correct head and tail type in the extracted entities
        head_found = [entity for entity, label in entities if label == head_entity]
        tail_found = [entity for entity, label in entities if label == tail_entity]

        if head_found and tail_found:
            for h in head_found:
                for t in tail_found:
                    for relation in valid_relations:
                        relations.append((h, t, relation_labels.get(relation, 'Unknown')))
    
    return relations

# Function to combine both NER and RE outputs dynamically
def get_named_entities_and_relations(sentence):
    print("=== Input Sentence ===")
    print(f"{sentence}\n")

    # Get named entities
    entities = get_named_entities(sentence)
    print("=== Named Entities ===")
    for entity, label in entities:
        print(f"  - {entity}: {label}")  # Display entity and its type

    # Get relations based on the rules
    relations = get_relations_from_rules(entities)
    print("\n=== Relations ===")
    if relations:
        for relation in relations:
            print(f"  - {relation[0]} --[{relation[2]}]--> {relation[1]}")
    else:
        print("  No relations found.")

# Generate a random test example and run inference
random_example = generate_random_test_example()

# Assuming 'tokens' is a list of words in your example
tokens = random_example[0].get('tokens', [])
sentence = ' '.join(tokens)  # Combine tokens into a sentence

if sentence:
    # Run the full inference pipeline
    get_named_entities_and_relations(sentence)
else:
    print("No tokens found in the random test example.")


```
